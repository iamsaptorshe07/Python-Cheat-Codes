```python
# Normal / Standard way of creating pandas dictionary manually using dictionary method
# Dictionary : It is a key value paird object where the key value represents the heading name of the column and a list
#              will be passed as the values of that header.
name=['Saptorshe Das','Samrat Basak','Gobinda Ghosh','Amar Sarkar','Sayan Naskar','Sayan Samanta','Sankha Mondal']
designation = ['Data Scientist','Accountant','Web Developer','Data Analyst','Cabin crew','Gov Sservice','Software Engineer']
salary=[128000,45000,45000,50000,40000,40000,48000]
creating_data_with_dictionary = {'Name':name,'Designation':designation,'Salary':salary}
creating_df = pd.DataFrame(creating_data_with_dictionary)# Passing the data with column header name
```

```python
# Printing the created dataframe
creating_df
```

```python
# dataframe.describe() ------> It prints the statistics of the dataframe on which column it is possible to perform
creating_df.describe()
```

```python
# Perform statistical operation on columns manually
minimum_value = creating_df['Salary'].min()  # --> Minimum value
maximum_value = creating_df['Salary'].max()
mean_value = creating_df['Salary'].mean()
standard_deviation = creating_df['Salary'].std()
```

```python
print(minimum_value,maximum_value,mean_value,standard_deviation)
```

```python
# Conditional selection like SQL query
# Printing the rows where salary is more than or equa to 48000
creating_df[creating_df.Salary>=48000]
```

```python
# Multiple conditional statement like SQL
# Print the details of the Employee who gets the highest salary
creating_df[creating_df['Salary']==creating_df.Salary.max()]
```

```python
# Print the details of the Employee who gets the lowest salary
creating_df[creating_df['Salary']==creating_df.Salary.min()]
```

```python
# Change the index name with any of column
new_data = creating_df.set_index('Name').copy()
new_data
```
