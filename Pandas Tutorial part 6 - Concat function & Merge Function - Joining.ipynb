#  Pandas Concat Function :
For more visit pandas official documentattion

```python
import pandas as pd
import numpy as np
```

```python
data_frame1 = pd.DataFrame({
    "City" : ['Mumbai','Kolkata','Bangalore'],
    "Temperature" : [32,30,31],
    "Humidity" : [80,60,78]
})
```

```python
data_frame1
```

```python
data_frame2 = pd.DataFrame({
     "City" : ['New York','Chicago','California'],
    "Temperature" : [22,14,26],
    "Humidity" : [40,30,28]
})
```

```python
data_frame2
```

```python
# Concatinating data frame row wise 
# Concat the above two dataframe
concatinated_df1 = pd.concat([data_frame1,data_frame2]) # uncontinous index
concatinated_df1
```

```python
concitanated_df2 = pd.concat([data_frame1,data_frame2],ignore_index=True) # Continious index value will be return
concitanated_df2
```

```python
# Concat data frames key / group wise
df = pd.concat([data_frame1,data_frame2],keys=['India','USA'])  # ignore_index argument will not work in this case
df
```

```python
# Get the information about particular index from the above concatinated data frame
print(df.loc['India'])
print()
print(df.loc['USA'])
```

```python
# Concatinating Data Frame column wise
data_frame3 = pd.DataFrame({
    "City" : ['Mumbai','Kolkata','Bangalore'],
    "Temperature" : [22,14,26],
    "Humidity" : [40,30,28]
})
```

```python
data_frame3
```

```python
data_frame4 = pd.DataFrame({
    "City" : ['Mumbai','Kolkata','Bangalore','Chennai'],
    "Wind Speed" : [22,14,26,80],
})
```

```python
data_frame4
```

```python
pd.concat([data_frame3,data_frame4],axis = 'columns')
```

# Pandas Merge Function

```python
# Merge function is quite simmiar to join function of SQL
df = pd.merge(data_frame3,data_frame4, on="City") # by default it takes how='inner'
df
# By deafult this method performs inter secion operation ( Inner Join)
```

```python
pd.merge(data_frame3,data_frame4,on='City',how='outer') # Performs outer join
```

```python
# Left join
pd.merge(data_frame3,data_frame4,on='City',how='left',indicator=True)
```

```python
pd.merge(data_frame3,data_frame4,on='City',how='right')
```
