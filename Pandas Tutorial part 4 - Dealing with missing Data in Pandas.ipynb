# How to deal with missing values
1 : using fillna method to fill the missing values using different ways <br>
2 : Interpollate to make a guess on missing values using interpolation <br>
3 : using dropna method to drop the rows with missing if it will not effect on the overall result

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
file = 'missingdata.csv'
df = pd.read_csv(file)
```

```python
# Converting the Day column data to a data format as it is in String format 
new_df = pd.read_csv('missingdata.csv',parse_dates=['Day'])
```

```python
type(new_df['Day'][0])
```

```python
# Setting the Day column as index
new_df.set_index('Day',inplace=True)
```

```python
new_df
```

```python
# Using method 1 : fillna() method
df = new_df.copy()
```

```python
df['Temparature'].fillna(df['Temparature'].mean(),inplace=True)
```

```python
df
```

```python
# Method 2 : Interpolation
print("We can perform linear interpolation method to fill the missing value")
in_df = new_df.copy()
```

```python
in_df
```

```python
in_df.interpolate(inplace=True)   # This is the best method to use to deal with missing values 
in_df
```

```python
# Method 3 : dropna method()  -------- It will drop the rows with missing values
# Before using this method we must consider the impact of dropping those rows in our overall result
new_df.dropna(inplace=True)
```
