2 : Interpollate to make a guess on missing values using interpolation <br>
3 : using dropna method to drop the rows with missing if it will not effect on the overall result

```python
# Converting the type of Day column data into date format type format as it is in String format 
new_df = pd.read_csv('missingdata.csv',parse_dates=['Day'])
```

```python
# Setting the Day column as index
new_df.set_index('Day',inplace=True)
```

```python
# Using method 1 : fillna() method
df = new_df.copy()
df.head(5)
```

```python
print(df['Temparature'].mean())
print(df['Temparature'].median())
print(df['Temparature'].mode())
```

```python
# It can be fill uped using some statistical methods like mean , median , mode which is suitable.
df['Temparature'].fillna(df['Temparature'].mean(),inplace=True) # Filling the temparature Nan values with the mean values of the given data set
```

```python
#df['Windipped'].fillna(df['Windipped'].mode(), inplace = True)
df
```

```python
# Filling the missing values with median value 
df['Windipped'].fillna(df['Windipped'].median(), inplace = True)
df
```

```python
df['Event'].fillna(df['Event'].mode(), inplace = True)
df
```

```python
# Carry forward the data of previous row
df2 = new_df.copy()
df2.head(5)
```

```python
# Copying the value from previous row
print(df2.fillna(method = 'ffill')) # ffill -- > Forward fill
print("\n \n \n")
# Copying the value from next row
print(df2.fillna(method='bfill')) #bfill -- > Back fill
print("\n \n \n")
#We can even pass axis as parameter
print(df2.fillna(method = 'ffill', axis='rows'))
print("\n \n ")

print(df2.fillna(method = 'ffill', axis='columns'))
# For more please visit pandas official documentation
```

```python
# Method 2 : Interpolation
print("We can perform linear interpolation method to fill the missing value")
in_df = new_df.copy()
```

```python
in_df
```

```python
in_df.interpolate(inplace=True)   # This is one of the best method to use to deal with missing values 
in_df # Visit the mathematical concept of interpollation, and for interpollate concept visit pandas interpollate documentation
```

```python
# Method 3 : dropna method()  -------- It will drop the rows with missing values
# Before using this method we must consider the impact of dropping those rows in our overall result
# Drop only those rows in which there is no value ( means all the values of row should be NaN)
new_df.dropna(how = 'all')
```

```python
# Drop only those rows in which row there is two non NaN value -- means if the row has atleast two non nan value it will kept that row
print(new_df.dropna(thresh = 2))
print("\n")
# Drop only those rows in which row there is one non NaN value 
print(new_df.dropna(thresh = 1))
```
