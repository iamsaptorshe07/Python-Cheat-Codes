```python
# Creating through list of tuples ---> here we passed the rows instead of column values unlike dictionary with lists
column_name = ['Name','Address','Contact No']
data = [
    ('Saptorshe Das','India',7278660111),
    ('Shiba Das','America',6290088603),
    ('Samrat Basak','London',7890813887),
    ('Chintu Basak','America',9230882388)
]
data_frame = pd.DataFrame(data,columns=column_name) # Here we must have to pass the column name serially
```

```python
data_frame
```

```python
# Creating dataframe using list of dictionaries  -- > Here we pass the column and row value together as a list, It is a non-efficienet way
vals = [
    {'Day':'01/12/2000','Age':20},
    {'Day':'11/10/2010','Age':30},
    {'Day':'21/01/2009','Age':15}
]
```

```python
df = pd.DataFrame(vals)
```
