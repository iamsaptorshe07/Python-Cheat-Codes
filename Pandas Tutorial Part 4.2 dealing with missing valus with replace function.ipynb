# Fill Missing values with replace function

```python
import pandas as pd
import numpy as np
```

```python
file = 'weather_data.csv'
df = pd.read_csv(file)
```

```python
# We can see there are some unusual values
df
```

```python
new_df = df.copy()
new_df.head(5)
```

```python
# To replace this unusual values we can use replace function  -- Syntax df.replace(value_which_we_want_to_replace, replacing_value)
df.replace(-99999,np.NaN)
```

```python
df['temperature'][3] = 98282
```

```python
df
```

```python
df.replace([-99999,98282],np.nan)
```

```python
#  By Passing a dictionary
df = df .replace({
    -99999:np.nan,
     '0' : "Sunny"
})
```

```python
df
```

```python
# Use regular expression concept to deal with repace function
```

```python
# Replace list of values with another list of values
df.replace(['Rain','Sunny','Snow'],[1,2,3,],inplace = True)
```
